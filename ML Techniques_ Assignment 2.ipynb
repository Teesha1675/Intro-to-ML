```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
"1."
```

```python
# Loading the dataset
data = pd.read_csv('dataset.csv')
```

```python
X = data[['x']]
y = data['y']
```

```python
# Defining degrees
degrees = [1, 2, 3, 4, 5]

for degree in degrees:
    poly_features = PolynomialFeatures(degree=degree)
    X_poly = poly_features.fit_transform(X)

    model = LinearRegression()
    model.fit(X_poly, y)
    y_pred = model.predict(X_poly)

    r2 = r2_score(y, y_pred)
    print(f"Degree {degree} R-squared: {r2}")
```

```python
""" Now here we can see that there are negligible changes in R- squared values
 for degree 3,4 and 5 and also taking high degree polynomial can lead to overfitting."
```

```python
" Hence best fit degree of polynomial is 3"
```

```python
# Loading the dataset
data2 = pd.read_csv('dataset2.csv')
```

```python
X_new = data2[['x']]
y_new = data2['y']
```

```python
# Linear Regression
linear_model = LinearRegression()
linear_model.fit(X_new, y_new)
linear_preds = linear_model.predict(X_new)
linear_mse = mean_squared_error(y_new, linear_preds)
```

```python
print(linear_mse)
```

```python
# Exponential Transformation
X_exp = np.exp(X_new)
exp_model = LinearRegression()
exp_model.fit(X_exp, y_new)
exp_preds = exp_model.predict(X_exp)
exp_mse = mean_squared_error(y_new, exp_preds)
```

```python
print(exp_mse)
```

```python
# Sine Transformation
X_sin = np.sin(X_new)
sin_model = LinearRegression()
sin_model.fit(X_sin, y_new)
sin_preds = sin_model.predict(X_sin)
sin_mse = mean_squared_error(y_new, sin_preds)
```

```python
print(sin_mse)
```

```python
# Logarithmic Transformation
X_log = np.log(X_new.values) if (X_new.values > 0).all() else np.log(X_new.values - min(X_new.values) + 1)  # Handling negative values or zeros if present
log_model = LinearRegression()
log_model.fit(X_log, y_new)
log_preds = log_model.predict(X_log)
log_mse = mean_squared_error(y_new, log_preds)
```

```python
print(log_mse)
```

```python
# Removing rows where 'X' column equals zero
X_no_zeros = data2[data2['x'] != 0]
# Extracting corresponding 'y' values for non-zero 'X' values
y_no_zeros = X_no_zeros['y']
```

```python
# Reciprocal Transformation
X_reciprocal = 1/ X_no_zeros
reciprocal_model = LinearRegression()
reciprocal_model.fit(X_reciprocal, y_no_zeros)
reciprocal_preds = reciprocal_model.predict(X_reciprocal)
reciprocal_mse = mean_squared_error(y_no_zeros, reciprocal_preds)
```

```python
print(reciprocal_mse)
```

```python
" The best-fitting function for the data is directly proportional to the minimum mean squared error. Hence sin function is the best to fit the data."
```
